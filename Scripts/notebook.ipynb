# Import Labraries and Data

```python
import pandas as pd


# Load the data
df = pd.read_csv('../Data/Input_FormResponse.csv')

df.head()
```

# Data Preparation

```python
# Drop the columns that are not needed and not labeled
df = df.drop(['Horodateur', 'Adresse e-mail', '9. Quelles améliorations aimeriez-vous voir sur le site ?'], axis=1)

df.head()
```

```python
# Rename the columns
df.columns = ["Age", "Sexe", "Statut", "Freq_Utilisation", "Facilite_Utilisation_note", "Design_note", "Contenu_note", "UX_mobile_note"]

df.head()
```

Forme disjonctif Z

```python
# Représenter la data sous forme disjonctif Z
df_disjointif = pd.get_dummies(df)

df_disjointif.head()
```

# Save the data

```python
# Save the cleaned data
df.to_csv('../Data/Output1_Cleaned_FormResponse_MatriceX.csv', index=False)
df_disjointif.to_csv('../Data/Output2_Disjointif_FormResponse_MatriceZ.csv', index=False)
```
